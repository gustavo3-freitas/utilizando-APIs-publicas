````python
# Função que recebe como argumento o UF da consulta e o tipo de interessado
def consulta_pvl(uf, tipo_interessado):
    base_url = f"http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl"
    params = {
        'uf': uf,
        'tipo_interessado': tipo_interessado
    }

    try:
        response = requests.get(base_url, params=params)
        data = response.json()
        df = pd.DataFrame(data['items'])
        return df
    except requests.exceptions.RequestException as e:
        print(f"Erro na consulta da API: {e}")
        return pd.DataFrame()  # Retorna um DataFrame vazio em caso de erro
````
